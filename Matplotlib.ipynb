```python
from numpy.random import randn
fig = plt.figure(figsize=(10,100));ax = fig.add_subplot(1,1,1)
```

```python
ax.plot(randn(1000).cumsum(), 'k', label='one', alpha = 0.5,  linewidth=0.3)
ax.plot(randn(1000).cumsum(), 'd-', label = 'two',  linewidth=0.3)
ax.plot(randn(1000).cumsum(), 'k.', label = 'three',  linewidth=0.3)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
```

```python
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color = 'k', alpha = 0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color = 'b', alpha = 0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color = 'g', alpha = 0.5)
```

```python
%matplotlib inline
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```
